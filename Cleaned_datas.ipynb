# Barcelona

```python
#Read CSV file
barcelona_listings = pd.read_csv('airbnb/barcelona_listings.csv')
barcelona_calendar = pd.read_csv('airbnb/barcelona_calendar.csv')
barcelona_reviews = pd.read_csv('airbnb/barcelona_reviews.csv')
```

```python
# Check id
print(barcelona_listings['id'].isnull().sum())

# Check date
print(barcelona_calendar['date'].isnull().sum())

# Check id
print(barcelona_reviews['id'].isnull().sum())
```

```python
# country
print(len(barcelona_listings_index_set['country']))
print(len(barcelona_listings_index_set.dropna(subset=['country'])))
print(barcelona_listings_index_set['country'].unique())
```
